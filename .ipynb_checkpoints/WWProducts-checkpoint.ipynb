# World Wide Products

```python
# For module loading issues
import sys
sys.path.append('/usr/local/lib/python3.8/site-packages')
from os import path
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from datetime import datetime
from fbprophet import Prophet
```

```python
products = pd.read_csv('data/raw/historical_product_demand.csv', parse_dates=['Date'])
products.head()
```

Ok, so we have a product code/category, a warehouse identifier a date of the product order and an order demand. Four of the five rows can be turned into integers which will help us later if we'd like to do some extensive model training. Integers allow quick parsing of data making it the perfect data type for training models. 

Let's clean up the data before we get into it. Order_Demand is *almost* perfect. Notice some of the Order_Demand rows have parentheses in them. Let's remove those because they are pointless. 

```python
products['Order_Demand'] = products['Order_Demand'].str.strip('()').astype(int)
```

Sweet. Now that Order_Demand is sorted out, let's think about forecasting. In order to deal with forecasting properly we'll need to use the Date field. I don't wanna sort through all the Date data for missing values, so let's go ahead and remove any product that has has no Date.

```python
products = products.query('not Date.isnull()')
```

Let's save this data as a processed csv file and then get into the feature engineering

```python
if not path.exists('data/processed/processed_products.csv'):
    products.to_csv('data/processed/processed_products.csv',index=False)
```

# Feature Engineering & Data Exploration
Sweet. Cleaning done. Let's jump into the data a little more and group it into different groups

```python
products['Month'] = products['Date'].apply(
    lambda dt: dt.replace(day=1)
)
data = products[['Product_Code', 'Month', 'Order_Demand']].groupby(['Product_Code', 'Month']).sum()
data.reset_index(level=['Product_Code', 'Month'], inplace=True)
data['Month_of_Year'] = data.apply(lambda r: r['Month'].month,axis=1)
```

The above allows us to group products into monthly increments. This will help organize our data in plots and graphs. Now let's get down to business and decide a specific product. The product we choose a very popular product because the more analysis we pull from the popular products, the more useful our analysis will be for businesses and marketing firms!

```python
total_count = data[['Product_Code', 'Order_Demand']].groupby('Product_Code').sum()
total_count.reset_index(level=['Product_Code'], inplace=True)
total_count.nlargest(10, 'Order_Demand').plot.bar(x='Product_Code', y='Order_Demand')
top_ten_codes = total_count.nlargest(10, 'Order_Demand')['Product_Code']
```

Boom. Top 10 popular products right there. Let's look into time data. Let's go ahead and grab the top ten Product Codes while we're at it

```python
def product_demand_over_time(product):
    p = data.query(f'Product_Code == "{product}"').sort_values('Month')
    f = plt.figure()
    axis = f.add_subplot(111)
    scatter = axis.hist(p['Month'], bins=len(p['Month']), weights=p['Order_Demand'])
    axis.set_title(f'{product.replace("_"," ")}: Demand over Time')
    axis.set_xlabel('Month')
    axis.set_ylabel('Demand')
    plt.show()
     
for p in list(top_ten_codes):    
    product_demand_over_time(p)
```

OK, for forecasting, it would be wise to choose a product that we could see being popular down the road. A good choice for that is a product with an upward trend. **Product 0083** is a good choice for this! Let's dive more into the the Month of the Year details to get a better idea of the long term results of each product, particularly 0083.

```python
def product_demand_over_months(product):
    p = data.query(f'Product_Code == "{product}"')
    f = plt.figure()
    axis = f.add_subplot(111)
    scatter = axis.hist(p['Month_of_Year'], bins=12, weights=p['Order_Demand'])
    axis.set_title(f'{product.replace("_"," ")}: Demand over Time')
    axis.set_xlabel('Month of the year')
    axis.set_ylabel('Demand')
    plt.show()
    
for p in list(top_ten_codes):
    product_demand_over_months(p)
```

# What's the Forecast?
It looks like **product 0083** has a slight increase in sales around the holidays. Likely a good Christmas gift! We have a product, now lets see what we can understand from the data.

To visualize our data, we'll look into Facebook's Prophet, data forecasting tool seen from class.

```python
my_product = 'Product_0083'
# In order to use Prophet(), we need to have ds and y column names
train = products.query(f'Product_Code == "{my_product}"')[['Date', 'Order_Demand']].rename(columns={'Date': 'ds', 'Order_Demand': 'y'})

fp = Prophet()
fp.fit(train)
```

```python
forecasting = fp.predict(fp.make_future_dataframe(periods=365))
figure = fp.plot(forecasting)
```

Ok, our results appear to be all over the map! The purely annual data is very cluttered here. We can see that there are a heavy number of orders around the holidays with an increase in product sales over time annually. 

This is good to know but, from our data analysis already, we already knew this. Let's try out a similar Prophet forecasting with our Monthly data! 

```python
train = data.query(f'Product_Code == "{my_product}"')[['Month', 'Order_Demand']]
# We need to conform our data to include ds and y column names 
train = train.rename(columns={'Month':'ds', 'Order_Demand':'y'})

fp = Prophet()
fp.fit(train)
```

```python
forecasting = fp.predict(fp.make_future_dataframe(periods=24, freq='M'))
figure = fp.plot(forecasting)
```

Wow, Prophet did some cool work here! We see a gradual increase with some severe dips in the beginning months of both 2017, 2018. This could be due to post-Christmas business blues. This high varience could be due to a number of factors. It's safe to assume our projections over time could dwindle in accuracy. Given we don't have any data points in 2018, 2019, the forecasting algorithm seems to settle for low points in the product's futures with a slight overall gradual increase.
